```python
import cv2
import torch
import matplotlib.pyplot as plt
import urllib.request
import numpy as np
```

```python
url = "https://images.unsplash.com/photo-1492724441997-5dc865305da7"
urllib.request.urlretrieve(url, "dark.jpg")

img = cv2.imread("dark.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Original Low-Light Image")
plt.axis('off')
```

```python

# Convert to LAB color space
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
l, a, b = cv2.split(lab)

# Apply CLAHE (adaptive AI-like enhancement)
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
cl = clahe.apply(l)

# Merge back
limg = cv2.merge((cl,a,b))
enhanced = cv2.cvtColor(limg, cv2.COLOR_LAB2BGR)
enhanced_rgb = cv2.cvtColor(enhanced, cv2.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Before")
plt.imshow(img_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("After (Enhanced)")
plt.imshow(enhanced_rgb)
plt.axis('off')

plt.show()
```

```python

# Simple contrast improvement check
before_mean = np.mean(img)
after_mean = np.mean(enhanced)

print("Brightness Before:", before_mean)
print("Brightness After:", after_mean)

improvement = ((after_mean - before_mean) / before_mean) * 100
print("Improvement %:", improvement)
```

```python
url2 = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/fruits.jpg"
urllib.request.urlretrieve(url2, "dark2.jpg")

img2 = cv2.imread("dark2.jpg")

# Same processing repeat
lab2 = cv2.cvtColor(img2, cv2.COLOR_BGR2LAB)
l2, a2, b2 = cv2.split(lab2)

cl2 = clahe.apply(l2)
limg2 = cv2.merge((cl2,a2,b2))
enhanced2 = cv2.cvtColor(limg2, cv2.COLOR_LAB2BGR)

plt.imshow(cv2.cvtColor(enhanced2, cv2.COLOR_BGR2RGB))
plt.title("Second Image Enhanced")
plt.axis('off')
```
